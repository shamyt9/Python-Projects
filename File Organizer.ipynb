```python
import os
import shutil

SOURCE_FOLDER=os.path.join("C:","Users","mrsha","Downloads")

# FILE TYPE CATEGORY

file_type={
     "Images": [".jpg", ".png", ".jpeg", ".gif",".webp"],
     "Videos": [".mp4", ".avi", ".mkv",".mov"],
     "Music": [".mp3", ".wav",".mpeg"],
     "PDF": [".pdf"],
     "Documents": [".docx", ".txt"],
     "Python": [".py"],
     "Softwares":[".exe"],
     "Compressed FILES":[".zip",".rar"],
    "Presentation":[".pptx",".ppt"]
}

# creaing folder if not exists
for folder in file_type:
    folder_path=os.path.join(SOURCE_FOLDER,folder)
    if not os.path.exists(folder_path):
        os.mkdir(folder_path)

# scanning the file
for file in os.listdir(SOURCE_FOLDER):
    file_path=os.path.join(SOURCE_FOLDER,file)
    if os.path.isfile(file_path):
        extension=os.path.splitext(file)[1].lower()
        for folder,ext_list in file_type.items():
            if extension in ext_list:
                shutil.move(file_path,os.path.join(SOURCE_FOLDER,folder,file))
                print(f"Moved {file} in {folder}")
                break
```
